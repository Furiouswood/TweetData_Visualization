## Tweet Target Analysis
Count how many tweets are targeted towards each category (Race, Religion, Gender, Sexual Orientation, Miscellaneous)

```python
# Count tweets targeted towards each category
# A tweet is "targeted" if it has a specific group (not the default "no target" value)

race_targeted = (data2["Race"] != "No_race").sum()
religion_targeted = (data2["Religion"] != "Nonreligious").sum()
gender_targeted = (data2["Gender"] != "No_gender").sum()
orientation_targeted = (data2["Sexual Orientation"] != "No_orientation").sum()
misc_targeted = data2["Miscellaneous"].notna().sum()

target_summary = pd.DataFrame({
    "Category": ["Race", "Religion", "Gender", "Sexual Orientation", "Miscellaneous"],
    "Tweet Count": [race_targeted, religion_targeted, gender_targeted, orientation_targeted, misc_targeted]
})
target_summary["Percentage (%)"] = (target_summary["Tweet Count"] / len(data2) * 100).round(2)

print(f"Total tweets: {len(data2)}\n")
print(target_summary.to_string(index=False))
```

```python
# Breakdown of specific groups within each category
print("=== Race Breakdown ===")
print(data2[data2["Race"] != "No_race"]["Race"].value_counts().to_string())

print("\n=== Religion Breakdown ===")
print(data2[data2["Religion"] != "Nonreligious"]["Religion"].value_counts().to_string())

print("\n=== Gender Breakdown ===")
print(data2[data2["Gender"] != "No_gender"]["Gender"].value_counts().to_string())

print("\n=== Sexual Orientation Breakdown ===")
print(data2[data2["Sexual Orientation"] != "No_orientation"]["Sexual Orientation"].value_counts().to_string())

print("\n=== Miscellaneous Breakdown ===")
print(data2[data2["Miscellaneous"].notna()]["Miscellaneous"].value_counts().to_string())
```
